# seleksi fitur

```python
features = ['weight', 'resoloution', 'ppi', 'cpu_core', 'cpu_freq', 'internal_mem', 'ram', 'RearCam', 'Front_Cam', 'battery', 'thickness']
x = df[features]
y = df['Price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
#'weight:150', 'resoloution:5', 'ppi:441', 'cpu_core:4', 'cpu_freq:2.3', 'internal_mem:16', 'ram:2', 'RearCam:13', 'Front_Cam:2', 'battery:2300', 'thickness:10'
input_data = np.array([[150, 5, 441, 4, 2.3, 16, 2, 13, 2, 2300, 10]])

prediction = lr.predict(input_data)
print('Estimasi Harga Telepon Seluler (IRR) :', prediction, 'IRR')
```
